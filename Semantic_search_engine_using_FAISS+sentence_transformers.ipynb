```python
# Generate Embeddings
from sentence_transformers import SentenceTransformer
import numpy as np

documents = [
    "How do I learn machine learning?",
    "What is the best way to learn Python?",
    "Tips to learn data science quickly.",
    "How does deep learning work?",
    "Best courses for AI and ML.",
    "What are embeddings in NLP?",
    "How does FAISS help in search?",
    "What is vector similarity search?",
    "How do transformers models work?",
    "Learn NLP with real projects."
]

model = SentenceTransformer('all-MiniLM-L6-v2')
document_embeddings = model.encode(documents, show_progress_bar = True)
```

```python
# Build FAISS Index
import faiss

embeddings_dim = document_embeddings.shape[1]

index = faiss.IndexFlatL2(embeddings_dim)

index.add(np.array(document_embeddings))
```

```python
# Search with Query
query = "I want to study artificial intelligence"
query_embedding = model.encode([query])

# search top 3 results
top_k = 3
distances, indices = index.search(query_embedding, top_k)

# show the results
print(f"Query : {query}")
print("\nTop Matches:")
for i in range(top_k):
    print(f"{i+1}. {documents[indices[0][i]]} (Distance: {distances[0][i]:.4f})")
```
